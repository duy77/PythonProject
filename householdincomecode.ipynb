```python

# Load the CSV file into a DataFrame
file_path = 'householdincome.csv'  
data = pd.read_csv(file_path)  

# Specify the correct column name for the income range
income_column = 'Estimate; Total: - $200,000 or more'

# Select the columns needed for analysis
columns = ['Id', 'Neighborhood', income_column]

# Find the 5 neighborhoods with the highest number of households in the income range $200,000 or more
top_5_neighborhoods = data.nlargest(5, income_column)[columns]

# Print the top 5 neighborhoods with the respective count of households
print("Top 5 neighborhoods with the most households earning $200,000 or more':")
print(top_5_neighborhoods)
```

```python
# Plot the data
import matplotlib.pyplot as plt

top_5_neighborhoods.plot.bar(x='Neighborhood', y='Estimate; Total: - $200,000 or more', rot=30)

plt.xlabel('Neighborhood')
plt.ylabel('Estimate Total Number')
plt.title('Top 5 Neighborhoods with the most earning $200,000 or more')
plt.show()
```

```python
# Load the CSV file into a DataFrame
file_path = 'householdincome.csv'  
data = pd.read_csv(file_path) 

# Specify the correct column name for the income range
income_column = 'Estimate; Total: - $200,000 or more'

# Select the columns needed for analysis
columns = ['Id', 'Neighborhood', income_column]

# Find the 5 neighborhoods with the most expensive houses
top_5 = data.nlargest(5,'Estimate; Total: - $200,000 or more' )[columns]

# Extract labels and sizes for the pie chart
labels = top_5['Neighborhood']
sizes = top_5['Estimate; Total: - $200,000 or more']

# Plotting the pie chart
plt.figure(figsize=(8, 8))  # Define the size of the pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Housing Values by City')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
